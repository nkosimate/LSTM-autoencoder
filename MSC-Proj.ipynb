## Feature Engineering

Rolling window is to be applied to the total active volume in the pit.
Drilling fluid volumes can fluctuate due to various operational activities like adding/removing fluid, changes in pump speed, etc. The rolling mean helps smooth out these short-term fluctuations and provides a clearer picture of the overall trend. A sudden and sustained increase in TVA could indicate a kick, where formation fluids enter the wellbore, displacing the drilling fluid. By monitoring the rolling mean, you can better detect such trends.

```python
#Adding rolling mean
df2['TVA_rolling_mean'] = df2[' TVA (SURFACE)'].rolling(window=60).mean()
# Fill NaN values created by rolling window
df2.fillna(method='bfill', inplace=True)
df2
```

```python
col_to_drop = [' TTV (SURFACE)',' TVT (SURFACE)',' MLOG LAG DEPTH',
               ' SPM3 (SURFACE)', ' SPM2 (SURFACE)',' SPM1 (SURFACE)',' TTV2 (SURFACE)']
df2 = df2.drop(columns=col_to_drop)
df2.head(5)
```

## On Bottom Drilling

We only need to check for the data when there is on bottom drilling thus, we need the bit depth and the depth to be the same and the ROP to be greater than zero.


```python
#only get data when it is on bottom drilling and that is when bit depth == depth
drilling_df = df2[df2[' BIT DEPTH (SURFACE)'] == df2[' DEPTH (SURFACE)']]
drilling_df = drilling_df[drilling_df[' ROP (SURFACE)'] > 0]
drilling_df = drilling_df.reset_index()
drilling_df = drilling_df.drop('index', axis=1)
drilling_df
```
